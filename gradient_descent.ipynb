```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset (study hours vs exam scores)
data = {
    "hours": np.array([1, 2, 3, 4, 5]),
    "scores": np.array([10, 20, 30, 40, 50])
}

# Extract independent (x) and dependent (y) variables
x = data["hours"]
y = data["scores"]

# Normalize the dataset for better convergence
x = (x - np.mean(x)) / np.std(x)

# Initialize parameters
m = 0  # Slope
b = 0  # Intercept

# Hyperparameters
learning_rate = 0.01
epochs = 1000

# Gradient Descent Algorithm with print statements
def gradient_descent(x, y, m, b, learning_rate, epochs):
    n = len(x)  # Number of data points
    costs = []

    for epoch in range(epochs):
        # Calculate predictions
        y_pred = m * x + b

        # Calculate gradients
        dm = (-2 / n) * np.sum(x * (y - y_pred))
        db = (-2 / n) * np.sum(y - y_pred)

        # Update parameters
        m -= learning_rate * dm
        b -= learning_rate * db

        # Compute the cost (Mean Squared Error)
        cost = np.mean((y - y_pred) ** 2)
        costs.append(cost)

        # Print the iteration, weight (m), bias (b), and cost
        if epoch % 100 == 0:  # Print every 100 iterations for clarity
            print(f"Iteration {epoch}: m = {m:.4f}, b = {b:.4f}, cost = {cost:.4f}")

    return m, b, costs

# Train the model
m, b, costs = gradient_descent(x, y, m, b, learning_rate, epochs)

# Print final results
print(f"Optimal slope (m): {m}")
print(f"Optimal intercept (b): {b}")
print(f"Final cost: {costs[-1]}")

# Plot the dataset and the best-fit line
plt.scatter(data["hours"], data["scores"], color="blue", label="Data Points")
plt.plot(data["hours"], m * ((data["hours"] - np.mean(data["hours"])) / np.std(data["hours"])) + b, color="red", label="Best-Fit Line")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Student Performance Prediction")
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent for Linear Regression
def gradient_descent(x, y):
    m = 0  # slope
    b = 0  # intercept
    n = len(x)  # number of data points

    learning_rate=0.01
    iterations=1000

    # Perform Gradient Descent
    for i in range(iterations):
        # Calculate predictions
        y_pred = m * x + b

        # Calculate the gradients
        dm = (-2/n) * np.sum(x * (y - y_pred))  # gradient for m
        db = (-2/n) * np.sum(y - y_pred)         # gradient for b

        # Update m and b
        m -= learning_rate * dm
        b -= learning_rate * db

        # Print iteration information
        if i % 100 == 0:  # Print every 100 iterations
            MSE = np.mean((y - y_pred)**2)  # Calculate Mean Squared Error (MSE) as the cost
            print(f"Iteration {i}: m = {m:.4f}, b = {b:.4f}, MSE = {MSE:.4f}")

    return m, b

# Generate some example data (you can replace this with real data)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Run Gradient Descent to find the best-fit line
m, b = gradient_descent(x, y)

# Print the results
print(f"Best-fit line: y = {m:.2f}x + {b:.2f}")

# Plot the results
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, m*x + b, color='red', label=f'y = {m:.2f}x + {b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
